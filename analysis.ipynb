#### 1) How did the prices of Mango (Raw-Ripe) vary over time in Nokha, Rohtas, Bihar? 
This question focuses on analyzing the price fluctuations of Mango (Raw-Ripe) in a specific market (Nokha, Rohtas, Bihar) throughout 2022. It aims to understand the temporal trends and identify any significant changes in prices.

```python
# Filter data for Nokha, Rohtas, Bihar
nokha_prices = data[(data['market'] == 'Nokha') & (data['district'] == 'Rohtas') & (data['state'] == 'Bihar')]
nokha_prices.head()
```

```python
# Plotting price variation over time
nokha_prices.plot(x='arrival_date', y='modal_price', kind='line',figsize=(10, 5))
plt.show()
```

### Conclusion
The prices of Mango (Raw-Ripe) in Nokha, Rohtas, Bihar showed decreasing trend over time.

#### 2) Did the prices of the Mango (Raw-Ripe) variety differ between Nokha, Rohtas, Bihar, and Vadhvan,Surendranagar, Gujarat?
This question compares the prices of Mango (Raw-Ripe) between two different markets, Nokha in Rohtas, Bihar, and Vadhvan,Surendranagar in Gujarat. It aims to explore whether there are regional variations in prices and understand the factors that might contribute to the differences.

```python
# Filter data for Mango (Raw-Ripe) in Nokha, Rohtas, Bihar
nokha_prices = data[(data['market'] == 'Nokha') & (data['district'] == 'Rohtas') & (data['state'] == 'Bihar') 
                    & (data['variety'] == 'Mango - Raw-Ripe')]

# Filter data for Mango (Raw-Ripe) in Vadhvan,Surendranagar, Gujarat
Vadhvan_prices = data[(data['market'] == 'Vadhvan') & (data['district'] == 'Surendranagar') & (data['state'] == 'Gujarat') 
                      & (data['variety'] == 'Mango - Raw-Ripe')]
# Compare prices
nokha_mean_price = nokha_prices['modal_price'].mean()
Vadhvan_mean_price = Vadhvan_prices['modal_price'].mean()

print("Mean price in Nokha, Rohtas, Bihar:", nokha_mean_price)
print("Mean price in Vadhvan, Surendranagar, Gujarat:", Vadhvan_mean_price)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
nokha_prices.plot(ax=axs[0])
Vadhvan_prices.plot(ax=axs[1])
plt.tight_layout()
plt.show()
```

### Conclusion
The mean price of Mango (Raw-Ripe) in Nokha, Rohtas, Bihar was 6250.0, while in Vadhvan, Surendranagar, Gujarat, it was 3069.4444444444443. This indicates that there may be regional differences in the prices of Mango (Raw-Ripe) between the two markets.

#### 3) Did the variety labeled as "Other" exhibit significant price fluctuations compared to Mango - Raw-Ripe in Ahmedabad, Gujarat? 
This question specifically focuses on comparing the price fluctuations of the Mango - Raw-Ripe variety with the "Other" variety in Ahmedabad, Gujarat. It aims to determine if there are any notable differences in price volatility between these two categories.

#### 4) How did the prices of Mango (Other) in Ahmedabad, Gujarat, change during the early part of the year (February-March 2022)? 
This question narrows down the analysis to a specific time period (February-March 2022) and examines the price trends of Mango (Other) in Ahmedabad, Gujarat during that period. It aims to uncover any seasonal patterns or price fluctuations during this specific timeframe.

```python
# Filter data for Mango (Other) in Ahmedabad, Gujarat
ahmedabad_prices = data[(data['market'] == 'Ahmedabad') & (data['district'] == 'Ahmedabad') &
                        (data['state'] == 'Gujarat') & (data['variety'] == 'Other')]

# Filter data for February-March 2022
start_date = pd.to_datetime('2022-02-01',infer_datetime_format=True)
end_date = pd.to_datetime('2022-03-31',infer_datetime_format=True)
arrival_date=pd.to_datetime(ahmedabad_prices['arrival_date'],infer_datetime_format=True)
feb_mar_prices = ahmedabad_prices[(arrival_date >= start_date) & (arrival_date <= end_date)]

# Calculate minimum, maximum, and average prices
min_price = feb_mar_prices['modal_price'].min()
max_price = feb_mar_prices['modal_price'].max()
avg_price = feb_mar_prices['modal_price'].mean()

print("Minimum price in February-March 2022:", min_price)
print("Maximum price in February-March 2022:", max_price)
print("Average price in February-March 2022:", avg_price)
```

```python
feb_mar_prices['modal_price'].plot()
plt.show()
```

### Conclusion
During the period of February-March 2022, the minimum price of Mango (Other) in Ahmedabad, Gujarat was 1750, the maximum price was 12500, and the average price was 6032.758620689655. This indicates the range and average level of prices during that specific time frame.
